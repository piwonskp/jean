```python
import seaborn as sns
import matplotlib.pyplot as plt

from ipynb.fs.full.jean.common import foreach_type, print_type
from ipynb.fs.full.jean.common._1load import appls
```

```python
def nulls_heatmap(df):
    _, ax = plt.subplots(figsize=(20, 5))
    sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis', ax=ax)

nulls_heatmap(appls)
```

```python
def titled_heatmap(df, t): print_type(t);nulls_heatmap(df);plt.show()

_ = list(foreach_type(appls, titled_heatmap))
```
