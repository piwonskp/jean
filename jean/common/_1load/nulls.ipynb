```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from ipynb.fs.full.jean.common._1load import appls

def nulls_heatmap(data, *args, **kwargs):
    sns.heatmap(data.isnull(), *args, cbar=False, yticklabels=False, cmap='viridis', **kwargs)
```

```python
_, ax = plt.subplots(figsize=(20, 5))
nulls_heatmap(appls, ax=ax)
```

```python
def nulls_facet_interface(*args, data, **kwargs): nulls_heatmap(data)


grid = sns.FacetGrid(appls, 'Rodzaj wniosku', dropna=False, height=4, aspect=4)
_ = grid.map_dataframe(nulls_facet_interface)
```
